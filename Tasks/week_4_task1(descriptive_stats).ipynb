```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from seaborn import load_dataset
```

# **loading data**

```python
data = sns.load_dataset("iris")
print("Iris Dataset:\n", data.head())

print("\nIris Dataset Shape:", data.shape)

#Check the columns
print("\nIris Dataset Columns:", data.columns)

#Check the empty data
print("\nIris Dataset Empty rows:\n", data.isnull().sum())
```

# **MEAN, MEDIAN,MODE.....**

```python
# Task 2: Calculate Mean, Median, Mode, Standard Deviation, Range, and Variance

# Mean of 'petal_length'
mean_petal_length = data['petal_length'].mean()
print(f"\nMean petal Length: {mean_petal_length:.3f}")

# Median of 'petal_length'
median_petal_length = data['petal_length'].median()
print(f"Median petal Length: {median_petal_length:.2f}")

# Mode of 'petal_length'
mode_petal_length = stats.mode(data['petal_length'])[0]
print(f"Mode petal Length: {mode_petal_length:.2f}")

# Range of 'petal_length'
range_petal_length = data['petal_length'].max() - data['petal_length'].min()
print(f"Range of petal Length: {range_petal_length:.2f}")

# Variance of 'petal_length'
variance_petal_length = data['petal_length'].var()
print(f"Variance of petal Length: {variance_petal_length:.2f}")

# Standard deviation of 'petal_length'
std_petal_length = data['petal_length'].std()
print(f"Standard Deviation of petal Length: {std_petal_length:.2f}")
```

# **Exercise 2: Visualizing Data Distributions with Histograms**

```python
# Task 1: Plot a Histogram of petal Lengths
plt.figure(figsize=(8, 6))
# Create the histogram with KDE
sns.histplot(data['petal_length'], bins=10, kde=True, color='blue', label='KDE Curve')
# Add a title and labels
plt.title("Distribution of petal Lengths", fontsize=16)
plt.xlabel("petal Length (cm)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
# Add a legend
plt.legend(title='Legend', loc='upper right')
# Save the figure
plt.savefig("petal_length_distribution.png", dpi=300, bbox_inches='tight')
# Show the plot
plt.show()
```

# **Exercise 3: Exploring Data Distributions**

```python
# Task 1: Explore Data Distribution for petal Length
# Skewness of 'petal_length'
petal_length_skewness = data['petal_length'].skew()
print("\nSkewness of petal Length:", petal_length_skewness)

# Checking if 'petal_length' is normally distributed
sns.histplot(data['petal_length'], kde=True)
plt.title("Normality Check of petal Length")
plt.show()
```
